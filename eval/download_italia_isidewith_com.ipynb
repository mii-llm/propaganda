```python
from bs4 import BeautifulSoup

from urllib.request import Request, urlopen
import re

req = Request('https://italia.isidewith.com/political-quiz', headers={'User-Agent': 'Mozilla/5.0'})
html_page = urlopen(req).read()

soup = BeautifulSoup(html_page, 'html.parser')
```

```python
print(soup)
```

```python
sections = soup.find_all('div', class_='sec_header_c')

#print(sections)

# Retrieve all h3 tags within this section
h3_tags = [section.find('h3') for section in sections]
#print(h3_tags[0:3])
#h3_tags = section.find_all('h3') if section else []

# Extract the text from each h3 tag
h3_values = [h3.get_text(strip=True) for h3 in h3_tags if h3]

# Print the extracted values
print(len(h3_values))
```

```python
for h3 in h3_values:
    print(h3)
```
